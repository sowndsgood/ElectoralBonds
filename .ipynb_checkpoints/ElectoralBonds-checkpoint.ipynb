##**Importing the files from drive.**





```python
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

encashment_pdf_path = '/content/drive/My Drive/ElectoralBonds/Encashment.pdf'
purchaser_pdf_path = '/content/drive/My Drive/ElectoralBonds/Purchaser.pdf'

import PyPDF2

def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text()
    return text

# Read content from Encashment.pdf
encashment_text = read_pdf(encashment_pdf_path)

# Read content from Purchaser.pdf
purchaser_text = read_pdf(purchaser_pdf_path)

'''
print("Text extracted from Encashment.pdf:")
print(encashment_text)

print("\nText extracted from Purchaser.pdf:")
print(purchaser_text)

'''
```

##**Top Political Parties based on number of denominations**

```python
from collections import Counter

# Preprocess text to extract political parties
def extract_political_parties(text):
    parties = []
    lines = text.split('\n')
    for line in lines[1:]:  # Skip the header line
        if line.strip():
            fields = line.split()
            if len(fields) >= 3:
                party = ' '.join(fields[1:-1]).strip()  # Extract the party name
                parties.append(party)
    return parties

# Read content from Encashment.pdf (assuming encashment_text contains the content)
encashment_political_parties = extract_political_parties(encashment_text)

# Calculate the top political parties
top_political_parties = Counter(encashment_political_parties).most_common(10)

# Print top political parties
print("Top Political Parties based on number of denominations:")
for party, count in top_political_parties:
    print(f"{party}: {count}")

import matplotlib.pyplot as plt

# Extracted top political parties and their counts
political_party_names = [party[0] for party in top_political_parties]
political_party_counts = [party[1] for party in top_political_parties]

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(political_party_names, political_party_counts, color='skyblue')
plt.ylabel('Frequency')
plt.xlabel('Political Party')
plt.title('Top Political Parties based on number of denominations')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()

```

##**Top Purchasers based on number of times purchased**

```python
from collections import Counter

# Preprocess text to extract purchasers
def extract_purchasers(text):
    purchasers = []
    lines = text.split('\n')
    for line in lines[1:]:  # Skip the header line
        if line.strip():
            fields = line.split()
            if len(fields) >= 3:
                purchaser = ' '.join(fields[1:-1]).strip()  # Extract the purchaser name
                purchasers.append(purchaser)
    return purchasers

# Read content from Purchaser.pdf (assuming purchaser_text contains the content)
purchaser_names = extract_purchasers(purchaser_text)

# Calculate the top purchasers
top_purchasers = Counter(purchaser_names).most_common(10)



# Print top purchasers
print("Top Purchasers based on number of times purchased")
for purchaser, count in top_purchasers:
    print(f"{purchaser}: {count}")


# Extracted top purchasers and their counts
purchaser_names = [purchaser[0] for purchaser in top_purchasers]
purchaser_counts = [purchaser[1] for purchaser in top_purchasers]

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(purchaser_names, purchaser_counts, color='salmon')
plt.ylabel('Frequency')
plt.xlabel('Purchaser')
plt.title('Top Purchasers based on number of times purchased')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

##**Top Political Parties based on Highest Denomination**

```python
import PyPDF2

def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text()
    return text

def extract_political_parties_and_denominations(text):
    parties_denominations = {}
    lines = text.split('\n')
    for line in lines[1:]:  # Skip the header line
        if line.strip() and line[0].isdigit():
            fields = line.split()
            if len(fields) >= 3:
                party = ' '.join(fields[1:-1]).strip()  # Extract the party name
                denomination = int(fields[-1].replace(',', ''))  # Extract the denomination
                parties_denominations[party] = parties_denominations.get(party, 0) + denomination
    return parties_denominations


# Read content from Encashment.pdf (assuming encashment_text contains the content)
encashment_text = read_pdf(encashment_pdf_path)

# Extract political parties and denominations
encashment_parties_denominations = extract_political_parties_and_denominations(encashment_text)

# Sort parties by total denomination in descending order
sorted_parties = sorted(encashment_parties_denominations.items(), key=lambda x: x[1], reverse=True)

# Extract top political parties
top_political_parties = sorted_parties[:10]  # Assuming you want top 10 parties

# Print top political parties with their total denomination
print("Top Political Parties based on Highest Denomination ")
for party, denomination in top_political_parties:
    print(f"{party}: {denomination}")

import matplotlib.pyplot as plt

# Extract top political parties and their denominations
top_parties = [party for party, denomination in top_political_parties]
denominations = [denomination for party, denomination in top_political_parties]

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(top_parties, denominations, color='yellow')
plt.xlabel('Total Denomination')
plt.ylabel('Political Party')
plt.title('Top Political Parties based on Highest Denomination')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest denomination on top
plt.tight_layout()
plt.show()
```

##**Number of Purchases by Year**

```python
import PyPDF2
import pandas as pd
import matplotlib.pyplot as plt

# Function to read text from PDF
def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text()
    return text



# Process the text to extract purchases by year
purchases_by_year = {}
lines = purchaser_text.split('\n')
for line in lines[1:]:  # Skip the header line
    if line.strip() and line[0].isdigit():
        fields = line.split()
        if len(fields) >= 3:
            purchase_date = fields[0].split('/')[-1]  # Extract the year from the purchase date
            year = int(purchase_date)
            purchases_by_year[year] = purchases_by_year.get(year, 0) + 1

print('Purchases basen on each year')
# Print the number of purchases for each year
for year in range(2019, 2025):
    print(f"Purchases in {year}: {purchases_by_year.get(year, 0)}")

# Plot the number of purchases for each year
years = list(range(2019, 2025))
purchases = [purchases_by_year.get(year, 0) for year in years]

plt.figure(figsize=(10, 6))
plt.bar(years, purchases, color='seagreen')
plt.xlabel('Year')
plt.ylabel('Number of Purchases')
plt.title('Number of Purchases by Year')
plt.xticks(years)  # Ensure all years are displayed on the x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
